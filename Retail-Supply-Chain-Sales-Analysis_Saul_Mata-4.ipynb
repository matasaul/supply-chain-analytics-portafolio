```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python

# Cargar el dataset
os.chdir("C:/Users/ADMIN/Desktop/saul")
df = pd.read_csv("Retail-Supply-Chain-Sales-Analysis.csv")
```

```python
# visualizacion
df.head(3)
```

```python
# Agrupar por segmento
segmentos = df.groupby("Segment")[["Sales","Profit"]].sum().sort_values(by="Profit", ascending=False)

segmentos
```

```python
import seaborn as sns

# Gráfico de barras: Ganancias por segmento
plt.figure(figsize=(8,6))
sns.barplot(x=segmentos.index, y=segmentos["Profit"], palette="Oranges")
plt.title("Rentabilidad por Segmento de Cliente")
plt.xlabel("Segmento")
plt.ylabel("Ganancias Total")
plt.show()
```

##### Conclusión 
El análisis muestra qué segmentos de clientes generan mayor volumen de ventas y mayor margen de ganancia.  
Esto permite a la empresa **priorizar estrategias de marketing y fidelización** en los segmentos más rentables,  
y diseñar planes específicos para mejorar la rentabilidad de los segmentos menos productivos.
