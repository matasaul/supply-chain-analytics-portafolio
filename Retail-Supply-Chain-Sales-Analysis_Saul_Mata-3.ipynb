```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python

# Cargar el dataset
os.chdir("C:/Users/ADMIN/Desktop/saul")
df = pd.read_csv("Retail-Supply-Chain-Sales-Analysis.csv")
```

```python
#  Explorar las columnas disponibles
df.head(3)
```

```python
# Agrupar por categoría
rentabilidad_categoria = df.groupby("Category")[["Sales", "Profit"]].sum().sort_values(by="Profit", ascending=False)
rentabilidad_categoria
```

```python
# Gráfico de barras de rentabilidad por categoría
rentabilidad_categoria["Profit"].plot(kind="bar", color="seagreen", figsize=(8,6))
plt.title("Rentabilidad por Categoría")
plt.ylabel("Profit")
plt.xlabel("Category")
plt.show()
```

##### Conclusión 
El análisis muestra qué categorías generan mayor margen de ganancia.  
Esto permite a la empresa **priorizar recursos y campañas** en las categorías más rentables,  
y evaluar estrategias de mejora en aquellas con menor margen.
